```python
#Q6:

# It looks like total_bedrooms, population, households, and median_income all have some semblance of a linear
# relationship with the target variable median_house_value
```

```python
#Q7:
df_bedrooms = df[['total_bedrooms', 'median_house_value']]
df_bedrooms = df_bedrooms.dropna()

bed = sp.stats.linregress(df_bedrooms['total_bedrooms'], df_bedrooms['median_house_value'])
pop = sp.stats.linregress(df['population'], df['median_house_value'])
hou = sp.stats.linregress(df['households'], df['median_house_value'])
inc = sp.stats.linregress(df['median_income'], df['median_house_value'])

print("total_bedrooms and median_house_value:\n", bed)
print("\n")
print("population and median_house_value:\n", pop)
print("\n")
print("households and median_house_value:\n", hou)
print("\n")
print("median_income and median_house_value:\n", inc)
```

```python
#Q8:

plt.figure(figsize=(12, 6))
plt.plot(df['total_bedrooms'], df['median_house_value'], 'o')           # scatter plot showing actual data
plt.plot(df['total_bedrooms'], bed.intercept + bed.slope*df['total_bedrooms'], 'r', label='fitted line')
plt.xlabel('total_bedrooms')
plt.ylabel('median_house_value')
plt.title('total_bedrooms vs median_house_value')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['population'], df['median_house_value'], 'o')           # scatter plot showing actual data
plt.plot(df['population'], pop.intercept + pop.slope*df['population'], 'r', label='fitted line')
plt.xlabel('population')
plt.ylabel('median_house_value')
plt.title('population vs median_house_value')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['households'], df['median_house_value'], 'o')           # scatter plot showing actual data
plt.plot(df['households'], hou.intercept + hou.slope*df['households'], 'r', label='fitted line')
plt.xlabel('households')
plt.ylabel('median_house_value')
plt.title('households vs median_house_value')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['median_income'], df['median_house_value'], 'o')           # scatter plot showing actual data
plt.plot(df['median_income'], inc.intercept + inc.slope*df['median_income'], 'r', label='fitted line')
plt.xlabel('median_income')
plt.ylabel('median_house_value')
plt.title('median_income vs median_house_value')
plt.legend()

plt.show()
```
