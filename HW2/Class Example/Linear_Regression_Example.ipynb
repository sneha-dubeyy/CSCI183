```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

```python
# Import and display first five rows of advertising dataset
advert = pd.read_csv('Advertising.csv')
advert = advert.drop(columns=['Unnamed: 0'])
advert.head()
```

```python
# Plot regression against actual data
plt.figure(figsize=(12, 6))
plt.plot(advert['TV'], advert['Sales'], 'o')           # scatter plot showing actual data
plt.xlabel('TV Advertising Costs')
plt.ylabel('Sales')
plt.title('TV vs Sales')

plt.show()
```

```python
# Build linear regression model using TV as predictor
# Split data into predictors X and output Y
predictors = ['TV']
X = advert[predictors]
y = advert['Sales']

# Initialise and fit model
lm = LinearRegression()
model = lm.fit(X, y)
```

```python
print(f'intercept = {model.intercept_}')
print(f'coefficient = {model.coef_}')
```

```python
pred_y = model.predict(X)
print(pred_y.shape)
print(pred_y)
```
